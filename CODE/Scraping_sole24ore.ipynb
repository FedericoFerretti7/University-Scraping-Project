## Cultura e tempo libero

```python
from bs4 import BeautifulSoup
import requests
import re

base_url = "https://lab24.ilsole24ore.com/qualita-della-vita/tabelle/2022/cultura-e-tempo-libero"
response = requests.get(base_url)
html_page = response.text

html_tree = BeautifulSoup(html_page)
```

```python
html_tree
```

```python
a=html_tree.find_all("script")
a
```

```python
b=str(a)
b2=b.replace('"',"")
b3=re.split(r'let datiTabella=', b2)[1]
b4=re.split(r'righe:', b3)[1]
b5=re.split(r',range:', b4)[0]
b5=b5.replace('\\u00ec', '')
b5_nome=re.findall("nome:[A-Z]\'?[A-Z]?[a-z]+\?\s?\-?[A-Z]?[a-z]+\-?[A-Z]?[a-z]+", b5)
b5_punteggio=re.findall("punti:\d+\.?\d+?", b5)
```

```python
lista2_punteggio=[]
for i in range(len(b5_punteggio)):
    lista2_punteggio.append(re.split(r':', b5_punteggio[i])[1])
    
lista2_punteggio
```

```python
lista2_nome=[]
for i in range(len(b5_nome)):
    lista2_nome.append(re.split(r':', b5_nome[i])[1])
    
lista2_nome
```

```python
import pandas as pd 
Cultura_e_tempolibero= pd.DataFrame({'Nome':lista2_nome, 'Cultura_e_tempolibero':lista2_punteggio})
Cultura_e_tempolibero
```

## Unisco

```python
import pandas as pd
together= pd.merge(Cultura_e_tempolibero, Affari_e_lavoro)
together1= pd.merge(together, Ricchezza_e_consumi)
together2= pd.merge(together1, Giustizia_e_sicurezza)
together3= pd.merge(together2, Demografia_e_societ)
together4= pd.merge(together3, Ambiente_e_servizi)
```

```python
together4
```

```python
together4["Cultura_e_tempolibero"]=pd.to_numeric(together4["Cultura_e_tempolibero"])
together4["Affari_e_lavoro"]=pd.to_numeric(together4["Affari_e_lavoro"])
together4["Ricchezza_e_consumi"]=pd.to_numeric(together4["Ricchezza_e_consumi"])
together4["Giustizia_e_sicurezza"]=pd.to_numeric(together4["Giustizia_e_sicurezza"])
together4["Demografia_e_societ"]=pd.to_numeric(together4["Demografia_e_societ"])
together4["Ambiente_e_servizi"]=pd.to_numeric(together4["Ambiente_e_servizi"])
```

```python
together4['Average_punt'] = together4[["Cultura_e_tempolibero","Affari_e_lavoro", "Ricchezza_e_consumi", "Giustizia_e_sicurezza", "Demografia_e_societ","Ambiente_e_servizi"]].mean(axis=1)
```

```python
together4
```

```python
together4= together4.sort_values("Average_punt", ascending=False)
together4
```

```python
together4.to_csv('QoL.csv', index=False)
```
