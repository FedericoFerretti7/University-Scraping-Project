```python
import requests
import json
import pandas as pd
import csv
import re

data=pd.read_csv("/content/Glas_Mod.csv")
data.drop(columns='Unnamed: 0')
```

```python
l=data[(data['Job_Location']!="Italy") & (data['Job_Location']!="Apulia") & (data['Job_Location']!="Remote") & (data['Job_Location']!="Lombardy")&(data['Job_Location']!="Veneto")&(data['Job_Location']!="Emilia-Romagna")&(data['Job_Location']!="30d+")]['Job_Location']
l
```

```python
l2=[]
for i in l:
    if i not in l2:
        l2.append(i)
l2
```

```python
base_url = f"https://api.opencagedata.com/geocode/v1/json?q=Brugine&key=a5b0d680bd0642a1b15239795b9a778f"
response = requests.get(base_url)
e=response.json()
c=e['results']
for i in c[0]['components']:
    #print(i)
    if (i== 'county') or (i == 'province'):
        print(c[0]['components'][i])
```

```python
Citta=[]
Provincia=[]
Regione=[]
for i in l2:
    base_url = f"https://api.opencagedata.com/geocode/v1/json?q={i}&key=a5b0d680bd0642a1b15239795b9a778f"
    response = requests.get(base_url)
    e=response.json()
    b=e["results"]
    if b!=[]: 
        for j in b[0]['components']:
        #print(i)
            if (j== 'county') or (j == 'province'):
            #print(j)
                Provincia.append(b[0]['components'][j])
            if (j=='state'):
                Regione.append(b[0]['components'][j])
                Citta.append(i)
    print(Regione)
```

```python
#Provincia Roma e inglese
for i in range(len(Provincia)):
    if Provincia[i] =="Roma Capitale":
        Provincia[i]=re.split(r" ", Provincia[i])[0]
    if Provincia[i] =="Monza and Brianza" :
        Provincia[i]=Provincia[i].replace("and", "-")
    if Provincia[i] =="Monza e Brianza" :
        Provincia[i]=Provincia[i].replace("e", "-")
    if Provincia[i] == "Mantua":
        Provincia[i]="Mantova"
    if Provincia[i] == "Provincia di Trento":
        Provincia[i]="Trento"
    if Provincia[i] == "Milan":
        Provincia[i]="Milano"
    if Provincia[i] == "Genoa":
        Provincia[i]="Genova"
    if Provincia[i] == "Aristanis/Oristano":
        Provincia[i]="Oristano"
    if Provincia[i] == "South Tyrol":
        Provincia[i]="Bolzano" 
```

```python
for i in range(len(Regione)):
    if Regione[i]=='Trentino-Alto Adige/Südtirol':
        Regione[i]='Trentino-Alto Adige'
```

```python
link= pd.DataFrame({'Città':Citta, 'Provincia':Provincia, 'Regione':Regione})
link
```
